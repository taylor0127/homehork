```python
import torch
import torchvision.transforms as transforms
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt

# 函數從網址加載圖片
def load_image(url):
    response = requests.get(url)
    return Image.open(BytesIO(response.content))

# 定義變換
transforms_dict = {
    "Original Image": None,
    "Resized": transforms.Resize((520, 1120)),
    "Center Cropped": transforms.CenterCrop(2000),
    "Color Jittered": transforms.ColorJitter(brightness=0.8, contrast=0.3, saturation=0.3),
    "Horizontally Flipped": transforms.RandomHorizontalFlip(p=1),
    "Randomly Rotated": transforms.RandomRotation(88)
}

# 加載圖片
img_url = 'https://cdn.gamma.app/emvg9zyulbdudkb/8770fa0cc2fa45188c75401431d4cabe/original/kbjfM3ZVTvFRyFkEGmzXNDPzSse.jpg'
img = load_image(img_url)

# 準備繪圖
fig, ax = plt.subplots(1, 6, figsize=(15, 5))  # 1 row，6 column

# 繪製和顯示變換後的圖像
for i, (title, transform) in enumerate(transforms_dict.items()):
    transformed_img = img
    if transform:
        transformed_img = transform(transformed_img)
    ax[i].imshow(transformed_img)
    ax[i].set_title(title)
    ax[i].axis('off')

plt.tight_layout()
plt.show()
```
